```python
import numpy as np
import pandas as pd
import ast
```

```python
movies= pd.read_csv("movies.csv")
credits= pd.read_csv("credits.csv")
```

```python
movies.head(1)
```

```python
# credits.head(1)['cast'].values
credits.head(1)
```

```python
movies= movies.merge(credits,on= 'title')
movies.head(1)
```

```python
movies.info()
```

```python
movies= movies[['movie_id','genres','title','overview','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast #converts a string containing a Python list - for this code
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
        
        
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords'] =movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter>2:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
        
```

```python
# take only top 3 cast members
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
#fetch director names from crew
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(1)
```

```python
# removing spaces between the words so there wont be data collision
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","")  for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x]) 
```

```python
movies.head(1)
```

```python
# make new column 'tags' which will contain all the information of genres, keywords, cast and crew columns
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
# lowercase the tags
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
# vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
# !pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
# stemming (action, actions) same word with different forms
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# compute the cosine similarity based on the vectors
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
recommend('Avatar')
```
